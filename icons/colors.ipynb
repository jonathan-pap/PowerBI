```python
"""
Simple Color Palette for Power BI Icons
========================================
Just the core colors with light/base/dark variants.
"""
import csv
import colorsys

OUTPUT = "color_palette.csv"


def hex_to_hsl(hex_color):
    h = hex_color.lstrip("#")
    r, g, b = [int(h[i:i+2], 16) / 255.0 for i in (0, 2, 4)]
    hue, l, s = colorsys.rgb_to_hls(r, g, b)
    return hue, s, l


def hsl_to_hex(h, s, l):
    r, g, b = colorsys.hls_to_rgb(h, l, s)
    return "#{:02X}{:02X}{:02X}".format(
        int(round(r * 255)), int(round(g * 255)), int(round(b * 255))
    )


def make_shade(hex_color, target_lightness):
    h, s, l = hex_to_hsl(hex_color)
    return hsl_to_hex(h, min(s, 1.0), target_lightness)


# ── Core colors ──────────────────────────────────────────────
COLORS = {
    "Red":      "#F44336",
    "Green":    "#4CAF50",
    "Blue":     "#2196F3",
    "Indigo":   "#3F51B5",
    "Orange":   "#FF9800",
    "Yellow":   "#FFEB3B",
    "Violet":   "#9C27B0",
    "Grey":     "#9E9E9E",
    "Maroon":   "#800000",
    "Black":    "#424242",
    "Olive":    "#808000",
    "Cyan":     "#00BCD4",
    "Pink":     "#E91E63",
    "Magenta":  "#FF00FF",
    "Tan":      "#D2B48C",
    "Teal":     "#009688",
    "White":    "#FFFFFF",
}

# ── Generate CSV ─────────────────────────────────────────────
rows = []
for name, base_hex in COLORS.items():
    light = make_shade(base_hex, 0.82)
    dark  = make_shade(base_hex, 0.25)

    for variant, hex_val in [("Light", light), ("Base", base_hex), ("Dark", dark)]:
        hc = hex_val.lstrip("#")
        rows.append({
            "ColorName":   name,
            "Variant":     variant,
            "Hex":         hex_val,
            "HexClean":    hc,
            "SvgSafe":     f"%23{hc}",
        })

with open(OUTPUT, "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["ColorName", "Variant", "Hex", "HexClean", "SvgSafe"])
    writer.writeheader()
    writer.writerows(rows)

print(f"Saved {len(rows)} colors → {OUTPUT}\n")
print(f"{'ColorName':<12} {'Variant':<8} {'Hex':<10}")
print(f"{'-'*12} {'-'*8} {'-'*10}")
for r in rows:
    print(f"{r['ColorName']:<12} {r['Variant']:<8} {r['Hex']:<10}")
```
